```python
import copy

n, m = map(int, input().split())
graph = []
result = 0
tmp = [[0]*m for _ in range(n)]
for _ in range(n):
    graph.append(list(map(int, input().split())))

#virus퍼트리기
def virus(x, y):
    nd = [(-1,0), (1,0), (0,-1), (0,1)]
    for i in range(4):
        nx = x+nd[i][0]
        ny = y+nd[i][1]
        if nx >= 0 and nx < n and ny >= 0 and ny < m:
            if tmp[nx][ny] == 0:
                tmp[nx][ny] = 2
                virus(nx, ny)
#안전영역 check
def check_safe():
    score = 0
    for i in range(n):
        for j in range(m):
            if tmp[i][j] == 0:
                score += 1
    return score

#dfs로 벽 세우기
def dfs(count):
    global result
    
    if count == 3:
        #virus 퍼트리기
        for i in range(n):
            for j in range(m):
                tmp[i][j] = graph[i][j]
        
        for i in range(n):
            for j in range(m):
                if tmp[i][j] == 2:
                    virus(i, j)
                    
        #안전영역 구하기
        result = max(result, check_safe())
        return
    
    #dfs로 벽 3개 세우기
    for i in range(n):
        for j in range(m):
            if graph[i][j] == 0:
                graph[i][j] = 1
                count += 1
                dfs(count)
                count -= 1
                graph[i][j] = 0
                
dfs(0)
print(result)                
```
