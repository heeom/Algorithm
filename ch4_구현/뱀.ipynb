```python
# 초기화
n = int(input())
board = [[0] * n for _ in range(n)] # 정사각보드
k = int(input())
# 사과위치입력
for _ in range(k):
    a, b = map(int, input().split())
    board[a][b] -= 1
    
L = int(input()) #방향변환 횟수
change = []
for _ in range(L):
    x, l = input().split()
    change.append((int(x), l))
change.reverse()

direction = [(0,1),(1,0),(0,-1),(-1,0)]
# 시작 : (0,0)에서 출발
r,c = 0,0
d = 0 #방향정보
time = -1 #걸린 시간
x, l = change.pop() # 변환 정보

while True : 
    time += 1
    # 현재 위치
    board[r][c] += 1
    #방향 변환 check
    if time == x: # 1<=X<=10000
        if l == 'D': #오른쪽으로 90도 회전
            if d < 3 :
                d += 1
            else:
                d = 0
        else:
            if d > 0:
                d -= 1
            else:
                d = 3
                
        if change:
            x, l = change.pop()
            
    # 움직이기
    nr = r + direction[d][0]
    nc = c + direction[d][1]
    print(time,':',nr, nc)
   
    #벽 check
    if (nc < 0 or nc >= n) or (nr < 0 or nr >= n):
        break
        
    #사과 유무 check     
    if board[nr][nc] == 1: #자기 몸
        break
    elif board[nr][nc] == -1: #사과 있
        board[nr][nc] += 1
    elif board[nr][nc] == 0: #사과 없
        board[r][c] -= 1

    r, c = nr, nc
    
print(time+1)
```

```python
# 초기화
n = int(input())
board = [[0] * (n+1) for _ in range(n+1)] # 정사각보드
k = int(input())
# 사과위치입력
for _ in range(k):
    a, b = map(int, input().split())
    board[a][b] -= 1
    
L = int(input()) #방향변환 횟수
change = []
for _ in range(L):
    x, l = input().split()
    change.append((int(x), l))
change.reverse()

direction = [(0,1),(1,0),(0,-1),(-1,0)]
# 시작 : (0,0)에서 출발
r,c = 1,1
d = 0 #방향정보
time = 0 #걸린 시간
x, l = change.pop()  # 변환 정보 1<=L<=100
snake = []
while True :
    # 현재 위치
    board[r][c] += 1
    snake.append((r,c))
    
    #방향 변환 check
    if time == x: # 1<=X<=10000
        if l == 'D':
            d += 1
        else:
            d -= 1
        if change:
            x, l = change.pop()
    
    time += 1

    # 움직이기
    nd = d % 4 
    if nd < -1 :
        nd = abs(nd)
    nr = r + direction[nd][0]
    nc = c + direction[nd][1]
   
    #벽 check
    if (nc < 1 or nc > n) or (nr < 1 or nr > n):
        break
    else:
        #사과 유무 check     
        if board[nr][nc] == 1: #자기 몸
            break
        elif board[nr][nc] == -1: #사과 있
            board[nr][nc] += 1
        elif board[nr][nc] == 0: #사과 없
            i, j = snake.pop(0)
            board[i][j] -= 1

    r, c = nr, nc
    
print(time)
```
